```python
import os
import pandas as pd

output_dir = 'forecasts_by_horizon'

def read_forecasts_by_model_type():
    statistical_forecasts = {}
    tree_forecasts = {}
    TimeGPT_forecasts = {}
    neural_forecasts = {}

    for file_name in os.listdir(output_dir):
        if file_name.endswith('.csv'):
            model_type = file_name.split('_forecasts_horizon_')[0]
            horizon = int(file_name.split('_horizon_')[0].split('.csv')[0])
            file_path = os.path.join(output_dir, file_name)
            df = pd.read_csv(file_path)

            if model_type == 'statistical':
                if horizon not in statistical_forecasts:
                    statistical_forecasts[horizon] = []
                statistical_forecasts[horizon].append(df)
            elif model_type == 'tree':
                if horizon not in tree_forecasts:
                    tree_forecasts[horizon] = []
                tree_forecasts[horizon].append(df)
            elif model_type == 'TimeGPT':
                if horizon not in TimeGPT_forecasts:
                    TimeGPT_forecasts[horizon] = []
                TimeGPT_forecasts[horizon].append(df)
            elif model_type == 'neural':
                if horizon not in neural_forecasts:
                    neural_forecasts[horizon] = []
                neural_forecasts[horizon].append(df)

    return {
        'statistical': statistical_forecasts,
        'tree': tree_forecasts,
        'TimeGPT': TimeGPT_forecasts,
        'neural': neural_forecasts
    }

# Example usage:
all_forecasts = read_forecasts_by_model_type()

# Now all_forecasts will contain a dictionary with keys 'statistical', 'tree', 'TimeGPT', and 'neural'
# Each key will have a nested dictionary where keys are horizons and values are lists of DataFrames.
# For example:
# all_forecasts['statistical'][1] will give you a list of DataFrames for statistical model with horizon 1.
```

```python
import plotly.graph_objects as go
import os
import plotly.io as pio

# Define paths for saving outputs
output_dir = 'outputs/'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Helper function to save plotly figures as HTML
def save_plotly_fig(fig, filename):
    pio.write_html(fig, os.path.join(output_dir, filename))

def plot_all_forecasts(df, forecasts_by_horizon):
    fig = go.Figure()

    # Add actual data
    fig.add_trace(go.Scatter(x=df['ds'], y=df['y'], mode='lines', name='Actual'))

    # Add forecasts
    for horizon, forecasts_list in forecasts_by_horizon.items():
        combined_forecasts = pd.concat(forecasts_list).reset_index(drop=True)
        for col in combined_forecasts.columns:
            if col in ['ds', 'unique_id', 'origin']:
                continue
            fig.add_trace(go.Scatter(x=combined_forecasts['ds'], y=combined_forecasts[col], mode='lines', name=f'{col} (Horizon {horizon})'))

    fig.update_layout(title='All Forecasts by Horizon', xaxis_title='Date', yaxis_title='Passengers')
    fig.show()
    save_plotly_fig(fig,"All Forecasts by Horizon")
```

```python

plot_all_forecasts(df, {**all_forecasts})
```
